Scaling

```python
x = df.iloc[:,2:].values
x
```

```python
scale = MinMaxScaler()
features = scale.fit_transform(x)
features
```

 Elbow Method


```python
inertia=[]
for i in range(1,16):
  k_means = KMeans(n_clusters = i)
  k_means=k_means.fit(features)
  inertia.append(k_means.inertia_)
```

```python
plt.figure(figsize=(20,7))
plt.subplot(1,2,1)
plt.plot(range(1,16),inertia,'bo-')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
```

```python
plt.subplot(1,2,2)
kmeans=KMeans()
visualize = KElbowVisualizer(kmeans,k=(1,16))
visualize.fit(features)
plt.suptitle("Elbow Graph and Elbow Visualizer")
visualize.poof()
plt.show()
```

Silhouettes Scores for each K value

```python
silhouette_avg = []
for i in range(2,16):
  kmeans = KMeans(n_clusters = i)
  cluster_labels = kmeans.fit_predict(features)
  silhouette_avg.append(silhouette_score(features,cluster_labels))
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(2,16),silhouette_avg,'rX-')
plt.title('Silhouette Score')
plt.xlabel('Values of K')
plt.ylabel('Silhouette Score')
```

K-means Model. Taking K value as 3 as per Elbow *Method*

```python
model = KMeans(n_clusters=3)
model = model.fit(features)
```

```python
y_km = model.predict(features)
centers = model.cluster_centers_
```

```python
df['Cluster'] = pd.DataFrame(y_km)
df.to_csv("Cluster Data",index=False)
```
